```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import special

chi = np.linspace(0,40,200000)
def Ylm(l):
    return special.sph_harm(0,l,0,0)

y_1 = 4*np.pi*((1/(4*np.pi)))**0.5*1j**0*special.j0(chi)*Ylm(0)
plt.plot(chi,y_1,c='black',label='$J_0(x)$')

y_2 = 4*np.pi*((3/4/np.pi))**0.5*1j**1*special.j1(chi)*Ylm(1)
plt.plot(chi,y_2,c='blue',label='$J_1(x)$')

y_3 = 4*np.pi*((5/4/np.pi))**0.5*1j**2*special.jv(2, chi)*Ylm(2)
plt.plot(chi,y_3,c='red',label='$J_2(x)$')

plt.legend()

plt.xlabel('$\chi$')
plt.ylabel('$y$')

plt.savefig('imaginary part plot')
```
